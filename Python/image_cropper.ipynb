```python
import cv2, os
import matplotlib.pyplot as plt
import random
```

```python
image_test = "Images/pool/testing/imagestest/imgtest2.png"
img = cv2.imread(image_test)
```

```python
def cv2_imshow(im, true_color = False):
    if true_color:
        im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    plt.figure(), plt.imshow(im), plt.axis('off')
```

```python
cv2_imshow(img, True)
print('height', len(img), "& width", len(img[0]))
print('height', len(img)//224, "& width", len(img[0])//224)
```

```python
size = 224*2
for i in range(0,len(img)//size):
    for j in  range(0,len(img[0])//size):
        img_tmp = cv2.resize(img[i*size:(i+1)*size,j*size:(j+1)*size,:], (224,224))
        cv2.imwrite(filename=f"Images/pool/testing/image_cropped/img_x{i}_y{j}.jpg",img=img_tmp)
```

```python
size = 224*2
for i in range(0,3):
    for j in  range(0,3):
        img_tmp = cv2.resize(img[i*size:(i+1)*size,j*size:(j+1)*size,:], (224,224))
        cv2_imshow(img_tmp, True)
```
